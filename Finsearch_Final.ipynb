```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.tsa.vector_ar.var_model import VAR

# Step 1: Data Collection
start_date = '2004-07-01'
end_date = '2024-07-01'

# S&P 500 Index Prices
sp500 = yf.download('^GSPC', start=start_date, end=end_date)

# Fed Funds Rate Approximation
fed_rates = yf.download('^IRX', start=start_date, end=end_date)

# S&P 500 E-mini Futures Prices
sp500_futures = yf.download('ES=F', start=start_date, end=end_date)

# Step 2: Data Preparation
# Align the dates and clean the data
data = pd.DataFrame({
    'SP500': sp500['Adj Close'],
    'Fed_Rates': fed_rates['Adj Close'],
    'SP500_Futures': sp500_futures['Adj Close']
}).dropna()

# Step 3: Exploratory Data Analysis (EDA)
# Plot the data
plt.figure(figsize=(14, 7))

plt.subplot(3, 1, 1)
plt.plot(data['SP500'], label='S&P 500 Index')
plt.title('S&P 500 Index Prices')
plt.legend()

plt.subplot(3, 1, 2)
plt.plot(data['Fed_Rates'], label='Fed Rates (Proxy)', color='orange')
plt.title('Fed Rates Approximation (13-Week T-Bill)')
plt.legend()

plt.subplot(3, 1, 3)
plt.plot(data['SP500_Futures'], label='S&P 500 Futures', color='green')
plt.title('S&P 500 Futures Prices')
plt.legend()

plt.tight_layout()
plt.show()

# Step 4: Econometric Analysis
# Simple Linear Regression to analyze the effect of Fed Rates on SP500 and Futures
# Independent variable (Fed Rates) and dependent variables (SP500 and Futures)
X = sm.add_constant(data['Fed_Rates'])

model_sp500 = sm.OLS(data['SP500'], X).fit()
model_futures = sm.OLS(data['SP500_Futures'], X).fit()

print("Regression Results - S&P 500")
print(model_sp500.summary())
print("\nRegression Results - S&P 500 Futures")
print(model_futures.summary())

# Vector Autoregression (VAR) Model to understand the interdependencies
model_var = VAR(data)
results_var = model_var.fit(maxlags=15, ic='aic')
results_var.summary()

# Impulse Response Function (IRF) to understand how a shock in Fed Rates affects the variables
irf = results_var.irf(10)
irf.plot(orth=False)
plt.show()

# Step 5: Interpretation
# Based on the results, you would interpret how changes in Fed Rates influence SP500 prices and Futures
# Analyzing coefficients and IRF plots for insights
```

```python
import pandas as pd
import yfinance as yf
import statsmodels.api as sm

# Define the time periods
periods = {
    '2004-2009': ('2004-07-01', '2009-06-30'),
    '2009-2014': ('2009-07-01', '2014-06-30'),
    '2014-2019': ('2014-07-01', '2019-06-30'),
    '2019-2024': ('2019-07-01', '2024-07-01'),
}

# Download the data from Yahoo Finance
data = yf.download(['^GSPC', 'SPY'], start='2004-07-01', end='2024-07-01', interval='1mo')
data['Fed_Rates'] = yf.download('^IRX', start='2004-07-01', end='2024-07-01', interval='1mo')['Adj Close']

# Loop through each period and run the regression
results = {}
for period, (start_date, end_date) in periods.items():
    period_data = data.loc[start_date:end_date].dropna()

    X = period_data[['Fed_Rates']]
    X = sm.add_constant(X)

    # S&P 500 Index - Use the correct column name from your DataFrame
    y_sp500 = period_data[('Adj Close', '^GSPC')]  # Adjust column name if needed
    model_sp500 = sm.OLS(y_sp500, X).fit()

    # S&P 500 Futures (approximated by SPY ETF) - Use the correct column name
    y_sp500_futures = period_data[('Adj Close', 'SPY')]  # Adjust column name if needed
    model_sp500_futures = sm.OLS(y_sp500_futures, X).fit()

    # Store the results
    results[period] = {
        'SP500': model_sp500.summary(),
        'SP500_Futures': model_sp500_futures.summary(),
    }

# Display the results
for period, result in results.items():
    print(f"--- {period} ---")
    print(result['SP500'])
    print(result['SP500_Futures'])
    print("\n\n")
```

```python
import pandas as pd
import yfinance as yf
import numpy as np
import statsmodels.api as sm

# Define the time period
start_date = '2004-07-01'
end_date = '2024-07-01'

# Download the data
data = yf.download(['^GSPC', '^IRX'], start=start_date, end=end_date, interval='1mo')

# Calculate monthly returns
data['SP500_Return'] = data['Adj Close']['^GSPC'].pct_change()
data['Risk_Free_Return'] = data['Adj Close']['^IRX'].pct_change()
data['Market_Return'] = data['SP500_Return']  # Market return is the same as SP500 return

# Drop NA values
data.dropna(inplace=True)

# Prepare data for CAPM model
X = data[['Market_Return']].copy()
X = sm.add_constant(X)  # Adds the intercept term
y = data['SP500_Return']

# Fit the CAPM model
model = sm.OLS(y, X).fit()

# Print the summary
print(model.summary())

# Extract Beta value
# Check the model summary to find the correct name for the market return coefficient
beta = model.params[1]  # Assuming 'Market_Return' is the second parameter after the constant
print(f'Beta of S&P 500 Index: {beta:.4f}')
```

```python
import pandas as pd
import yfinance as yf
import statsmodels.api as sm

# Define the time period
start_date = '2004-07-01'
end_date = '2024-07-01'

# Download historical data for Apple stock and the S&P 500 Index
data = yf.download(['AAPL', '^GSPC', '^IRX'], start=start_date, end=end_date, interval='1mo')

# Calculate monthly returns
data['AAPL_Return'] = data['Adj Close']['AAPL'].pct_change()
data['Market_Return'] = data['Adj Close']['^GSPC'].pct_change()
data['Risk_Free_Return'] = data['Adj Close']['^IRX'].pct_change()

# Drop NA values
data.dropna(inplace=True)

# Prepare data for CAPM model
X = data[['Market_Return']].copy()
X = sm.add_constant(X)  # Adds the intercept term
y = data['AAPL_Return']

# Fit the CAPM model
model = sm.OLS(y, X).fit()

# Print the summary
print(model.summary())

# Extract Beta value
# Use the correct key from the model summary to extract the Beta value
beta = model.params[1]  # Assuming 'Market_Return' is the second parameter after the constant
print(f'Beta of Apple Inc. (AAPL): {beta:.4f}')
```

```python


# Define the timeframe
start_date = "2004-07-01"
end_date = "2024-07-01"

# Download VIX and SPY (S&P 500 futures proxy) data
vix_data = yf.download("^VIX", start=start_date, end=end_date)
spy_data = yf.download("SPY", start=start_date, end=end_date)

# Plot VIX and S&P 500 futures on the same graph
fig, ax1 = plt.subplots(figsize=(12, 6))

color = 'tab:blue'
ax1.set_xlabel('Date')
ax1.set_ylabel('VIX Value', color=color)
ax1.plot(vix_data.index, vix_data['Close'], color=color, label="VIX")
ax1.tick_params(axis='y', labelcolor=color)
ax1.grid(True)

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

color = 'tab:red'
ax2.set_ylabel('S&P 500 Futures Price', color=color)  # we already handled the x-label with ax1
ax2.plot(spy_data.index, spy_data['Adj Close'], color=color, label="S&P 500 Futures (SPY)")
ax2.tick_params(axis='y', labelcolor=color)

fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.title("VIX and S&P 500 Futures (2004-2024)")
plt.show()
```

```python


# Calculate the correlation
correlation = vix_data['Close'].corr(spy_data['Adj Close'])
print(f"Correlation between VIX and S&P 500 Futures (SPY): {correlation:.4f}")
```

```python


# Calculate rolling correlations
rolling_corr = vix_data['Close'].rolling(window=60).corr(spy_data['Adj Close'])  # 60-day rolling window

# Find the date with maximum correlation
max_corr_date = rolling_corr.idxmax()
max_corr_value = rolling_corr.max()

print(f"Date of maximum correlation: {max_corr_date}")
print(f"Maximum correlation value: {max_corr_value:.4f}")

# Plot rolling correlations
plt.figure(figsize=(12, 6))
plt.plot(rolling_corr.index, rolling_corr.values, label='Rolling 60-day Correlation')
plt.axvline(max_corr_date, color='red', linestyle='--', label=f'Max Corr: {max_corr_value:.4f}')
plt.title('Rolling Correlation between VIX and S&P 500 Futures')
plt.xlabel('Date')
plt.ylabel('Correlation')
plt.legend()
plt.grid(True)
plt.show()
```
